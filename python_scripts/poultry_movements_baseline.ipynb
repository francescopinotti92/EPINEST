The function `generate_configuration_files` takes `simulation_settings` as an input and parses it.

```python
#========== Parse master file ==========#

baseArgs = generate_configuration_files( simulation_settings )

print( '# of configurations:', len( baseArgs ) )
```

`baseArgs` is a list of dictionaries whose individual elements are dictionaries containing specifications to generate simulation files for a single scenario. The previous step is necessary since `simulation_settings` may contain specifications to generate multiple scenarios (not explored here).

Dictionaries in `baseArgs` can then be passed to `setup_single_scenario_Bangladesh_2`.

This function finally generates files required for simulation.

```python
#========== Generate individual scenarios files ==========#

thepath = os.path.join( os.getcwd(), 'baseline_scenario_files' )

for i, tmp in enumerate( baseArgs ):
    setup_single_scenario_Bangladesh_2( tmp, i, os.path.join( thepath, 'Config' ),  
                                     os.path.join( thepath, 'Output' ) )
```

### Parsing transactions

Each row in the dataframe represents a single transaction.
Attributes of each transaction are:

- **transact**: a single string containing all info about the transaction
- **srcType**: type of seller (PF=farm,MM=middleman,V=vendor,M=market)
- **dstType**: type of buyer (PF=farm,MM=middleman,V=vendor,M=market)
- **srcId**: numeric ID of seller
- **dstId**: numeric ID of buyer
- **nb**: number of birds sold in this transaction
- **t**: time step of transaction
- **day**: day of transaction
- **wday**: working day of transaction

### Plotting

The following code makes some basic plots using individual transactions

```python
# plot number of birds sold daily by farms
df.loc[df['srcType'] == 'PF'].groupby( 'day' )['nb'].sum().plot( 
    ylabel = 'Birds sold by farms', xlabel = 'Time (day)' )
```

```python
# plot number of birds sold daily by middlemen
df.loc[df['srcType'] == 'MM'].groupby( 'day' )['nb'].sum().plot( 
    ylabel = 'Birds sold by middlemen', xlabel = 'Time (day)' )
```

```python
# plot the (cumulative) size distribution of transactions involving middlemen and vendors

df1 = df.loc[df['srcType'] == 'MM'].copy() # select appropriate transactions
sz_hist = get_counts( df1['nb'] )

sz_hist.cumsum().plot( ylabel = 'Cumulative distribution', xlabel = 'Time (day)',
                     label = 'Size distribution')

# there are some transactions involving very few chickens
# however, these chickens represent only a tiny portion of
# chickens sold in ALL transactions.
# To see this, plot the (cumulative) biased size distribution,
# which represents the fraction of chickens involved in transactions
# of size n or smaller

sz_hist_biased = sz_hist * sz_hist.index
sz_hist_biased = sz_hist_biased / sz_hist_biased.sum()

sz_hist_biased.cumsum().plot( label = 'Biased size distribution')

plt.legend()
```

## Plot farm layout

```python
# read geography
which_area = 'Dhaka'

gdf = gpd.read_file( './data/bangladesh_shp' )
gf = gdf.copy()
gf = gf.to_crs( epsg = 3106 )
```

```python
fig, ax = plt.subplots()

# plot the geography

gf.plot( ax = ax, facecolor = 'white', edgecolor = 'black', linewidth = 0.05 )

# plot farms

path_farms = './baseline_scenario_files/Config/farm_info_0_0.json'
farm_pts= []
with open( path_farms ) as readFile:
    farmData = json.load(readFile)
    for key, farm in farmData.items():
        x, y = farm['x'], farm['y']
        farm_pts.append( Point( x, y ) )
gfarm = gpd.GeoDataFrame( { 'geometry': farm_pts } )
gfarm.plot( ax = ax, color = 'slateblue', markersize = 0.75, marker = 'x', label = 'Farm' )

# N.B. areas without farms are not included in the study

# indicate Dhaka

lat_dhaka = 23.777176
lon_dhaka = 90.399452
pDhaka = gpd.GeoDataFrame( { 'geometry': [Point( lon_dhaka, lat_dhaka )] } ).set_crs( epsg = 4326).to_crs( epsg = 3106 )
pt = pDhaka.geometry.values[0]
ax.annotate( 'Dhaka', (pt.x, pt.y), (pt.x + 5e4, pt.y - 1e4), fontsize = 16,
             arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3") )

ax.axis('off')
ax.set_xlim( 4.5e5, 6.5e5 )
ax.set_ylim( 2.6e6, 2.8e6 )
```
