# Explanation

This notebook shows how to change the inter-market mobility network. Nodes of this network represent markets, whereas a directed edge from market $A$ to market $B$ means that vendors can purchase chickens in $A$ and sell them in $B$.

First, we provide a description of the possible network topologies that are currently supported by our framework.

## Network model

The following Python code generates directed networks with `n_markets`. The network is created by adding one node at a time and connecting this node to previous nodes. Two parameters affect the shape of the network:

- `edge_density`: affects the number of connections generated.
- `prob_random_edge`: probability that an edge between a new node and an existing node is thrown at random rather then proportionally to the out-degree (outward arrows) of the latter node.

Increasing `edge_density` results in denser networks, while increasing `prob_random_edge` makes connections more random.

```python
import matplotlib.pyplot as plt
import networkx as nx
from popgenerator import generate_mixed_attachment_network, compute_weights_dirichlet
```

```python
n_markets = 20
edge_density = 0.1


fig, ax = plt.subplots( ncols = 2, figsize = ( 10, 4 ) )

# create two networks with 10% density of edges and varying levels of random connections
for i, prob_random_edge in enumerate( [ 0.001, 1. ] ):
    
    axi = ax[i]
    
    edge_list = generate_mixed_attachment_network( n_markets, edge_density, prob_random_edge )

    G = nx.DiGraph()
    G.add_nodes_from( range( n_markets ) )
    G.add_edges_from( edge_list )



    pos = nx.spring_layout(G, k = 2, iterations = 20)
    nx.draw_networkx(G, with_labels = False,
                     nodelist = list(G.nodes()),
                     pos = pos,
                     node_color = 'teal',
                     ax = axi)
    
    ax[i].set_title('edge density = {}, pr. random edge = {}'.format( edge_density, 
                                                                     prob_random_edge))
```

Network market parameters are specified in the **market_network_params** block. These parameters are:

- **mode**: type of network model. Now set to 'generative_mixed_attachment_model'.
- **rho_market**: edge density. Now set to 0.1. 
- **p_random_market**: probability of random connections. Now set to 0.01 (hieararchical network).
- **self_w_market**: edge weight parameter. It's the probability that a vendor buys and sell in the same market. Now set to 0.8, meaning that 20% of vendors move between markets.
- **het_w_market**: edge weight parameter. Modifies heterogeneity of travel probabilities. Now set to infinity so that all edges leaving one market have the same weight (vendors are equally likely to choose any selling destination given a source market).

The file `bird_mixing_market_0_0_0.txt` contains info about the origins of chickens found in each market at specific time points. Columns are:

- **t**: sampling time.
- **market**: identifier of market.
- **farm**: identifier of farm where chickens have been produced.

```python
path_mixing =  os.path.join( path_output, 'bird_mixing_market_0_0_0.txt' )
df = pd.read_csv( path_mixing, names = ['t','market','farm','nb'] )
df.head()
```

Using this information, we can quantify how likely it is to find chickens from different areas within the same market.

First, we need to assign each farm to its geographic area.

```python
# obtain a dictionary mapping each farm to the area it belongs to

path_farms = './market_scenario_files/Config/farm_info_0_0.json'

with open( path_farms ) as readFile:
    farmData = json.load( readFile )
    
farm2area = { el['id']: el['catchment_area'] for el in farmData.values() }


# perform the mapping (a new column 'area' should appear)
df['area'] = df['farm'].map( farm2area )

df.head()
```

Finally we calculate the number of markets in which it is possible find chickens from pairs of distinct areas.

This should result in a 50x50 matrix (there are 50 areas) whose entries range from 0 (no markets in common) to 20 (maximum number of markets).

The code below calculates the co-presence matrix and plots the distribution of the number of common markets for each pair of distinct areas.

```python
co = get_copresence_from_df( df, 'area', 'market', 'nb' ).tolist()
co = np.array( co )

# remember to remove diagonal entries (which have value 20 by definition)
plt.plot( np.bincount( co[~np.eye( len( co ) ).astype( bool ) ], minlength = 20 ) )
plt.xlabel( 'Number of markets in common' )
plt.ylabel( 'Frequency' )
```
